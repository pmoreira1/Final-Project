```python
%%time
from functions import functions as fc
import pickle
import pandas as pd
bins = [0, 0.80, 0.90, 1]
labels = [1, 2, 3]
origin = 235
destination = 111
loaded_model = pickle.load(open('models/'+str(origin), 'rb'))
hotels = pd.DataFrame(fc.db.select(query="SELECT * from business where category=2 and country='"+str(destination)+"'", all=True))
hotels_to_predict = hotels[['country', 'rating_cal']]
hotels_to_predict['b_review'] = pd.cut(hotels_to_predict['rating_cal'], bins, labels=labels)
hotels_to_predict.drop('rating_cal', axis=1, inplace=True)
hotels['match_level'] = loaded_model.predict(hotels_to_predict)
hotels.sort_values(by=['match_level','rating','review_count'], ascending=False)
```
